# Buffon's needle problem/Monte Carlo method to approximate $\pi$

## Explanation
This [video](https://www.youtube.com/watch?v=KSrPJe7y9oA) by MIT OCW explain the experiment and the Monte Carlo methods, and also the derivation of the solution to the problem.
The solution to the problem is as the following :
$$P(\text{The needle intersect with a line}) = \frac{2l}{\pi d}$$
Where :
- $l$ is the length of the needle;
- $d$ is the distance between the parallel lines.

## Experimental Setup
- I experimented with $l=7.4 \text{ cm}, d = 20.1 \text{ cm}$. We will approximate this propabilty as follow :
$$P(\text{The needle intersect with a line}) \approx \frac{\text{How many time there is an intersection}}{\text{Total number of essays}}$$


- I could draw parallel lines on a paper, but I preferred to use floor tiles instead to have a satisfying parralel lines.

## Downloading and visualizing the data
The observations of the experiment are registered in the csv file `buffonNeedle.csv`.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('buffonNeedle.csv')
df = pd.to_numeric(df["frequency"].apply(pd.eval))
l, d = 7.4, 20.1
df = df.apply(lambda x : 2*l/(x*d))
print(df.shape)
```

```python
plt.figure(figsize=(30, 10))
df.plot(linestyle="--", marker="o", markersize=4)
plt.axhline(np.pi, color="red", linestyle="--", label="$y = \pi$")
plt.legend()
plt.show()
```
